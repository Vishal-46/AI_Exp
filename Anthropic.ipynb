```python
#API client using Anthropic Package
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {"role": "user",
         "content": [{"type":"text", "text":"What is quantum computing?"}]}
    ]
)
```
